```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

DATA_DIR = Path("../data/raw")
# plt.style.use('default')
```

```python
# A month loading
df_jan = pd.read_parquet(DATA_DIR / "yellow_tripdata_2023-01.parquet")
print(f"Shape: {df_jan.shape}")
df_jan.head()
```

```python
neg_total_amount = len(df_jan[df_jan["total_amount"] < 0])
df_jan.shape[0]
print(f'{neg_total_amount} negative rows in the total_amount column')
print(f'total_amount: {neg_total_amount / df_jan.shape[0] * 100:.2f}%')
```

```python
neg_trip_distance = len(df_jan[df_jan['trip_distance'] < 0])
df_jan.shape[0]
print(f'{neg_trip_distance} negative rows are in the trip_distance column')
print(f'total_amount: {neg_trip_distance / df_jan.shape[0] * 100:.2f}%')
```

```python
(df_jan['tpep_pickup_datetime'] > df_jan['tpep_dropoff_datetime']).value_counts()
print("In 3 rows there are negative duration time")
df_jan[df_jan['tpep_pickup_datetime'] > df_jan['tpep_dropoff_datetime']]
```

```python
print(f"Zero passenger: {(df_jan['passenger_count'] == 0).sum()}")
print(f"NaN passenger:  {df_jan['passenger_count'].isnull().sum()}")
print(f"Total problems: {((df_jan['passenger_count'] == 0) | df_jan['passenger_count'].isnull()).sum()}")
```
